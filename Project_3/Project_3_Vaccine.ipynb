```python
import pandas as pd
from pandas import DataFrame
```

### Extracao e Tratamento de Dados

```python
def buildKey(df_item):
    dose = "1"
    #Se conter o valor 2 na string de dose alterar o valor para 2
    if('2' in df_item['vacina_descricao_dose']):
        dose = '2'
    #Cria um id para [uf, dose, mes]
    return "{}_{}_{}".format(df_item['estabelecimento_uf'], dose, df_item['vacina_mes_aplicacao'])

def addItemsFromDataframeDict(curr_dict, df_chunk):
    for _index, df_item in df_chunk.iterrows():
        #Chama o metodo para criar o ID
        key = buildKey(df_item)
        #Verifica se o id esta dentro do dicionario
        if (not key in curr_dict):
            #Se nao estiver add o valor do item para ID criado
            curr_dict[key] = df_item['document_id']
        else:
            #Se ja estiver soma o valor ao item que ja tem
           curr_dict[key] += df_item['document_id']
    return curr_dict
```

```python
def buildKeyVaccine(df_item):
    #Cria um id para [uf, dose, mes]
    return "{}_{}_{}".format(df_item['estabelecimento_uf'], df_item['nome_vacina'].upper(), df_item['vacina_mes_aplicacao'])

def getVaccineName(value):
    if('ASTRAZENECA' in value or 'COVISHIELD' in value):
        return 'ASTRAZENECA'
    if('CORONAVAC' in value):
        return 'CORONAVAC'
    if('PFIZER' in value or 'PENDENTE IDENTIFICAÇÃO' in value):
        return 'PFIZER'
    if('JANSSEN' in value):
        return 'JANSSEN'
    return value

def addItemsFromDataframeDictVaccine(curr_dict, df_chunk):
    for _index, df_item in df_chunk.iterrows():
        #Chama o metodo para criar o ID
        key = buildKeyVaccine(df_item)
        #Verifica se o id esta dentro do dicionario
        if (not key in curr_dict):
            #Se nao estiver add o valor do item para ID criado
            curr_dict[key] = df_item['document_id']
        else:
            #Se ja estiver soma o valor ao item que ja tem
           curr_dict[key] += df_item['document_id']
    return curr_dict
```

```python
#Definindo colunas que serao utilizadas da planilha
columns=['document_id', 'vacina_dataaplicacao', 'estabelecimento_uf', 'vacina_descricao_dose', 'vacina_nome']
#Colunas utilizada no group by
columns_group_by = ['estabelecimento_uf', 'vacina_descricao_dose','vacina_mes_aplicacao']
#Declarando dicionario para consolidacao dos dados
brasil_data = {}
vaccinebrasil_data = {}
#declaracao do chunksize
chunksize = 100000


# count_test = 0
#, error_bad_lines=False, engine="python", encoding='utf-8'
for chunk in pd.read_csv('./part-00000-c33feaf7-dccc-49dd-bb46-748542a80723-c000.csv', sep=';', chunksize=chunksize, usecols=columns):
    print("Iniciando um chunk de", chunksize)
    
    #Cria uma nova coluna com a data de vacinacao para o padrao YYYY-MM formatada
    chunk['vacina_mes_aplicacao'] = pd.to_datetime(chunk.vacina_dataaplicacao).dt.strftime('%Y-%m')
    chunk['nome_vacina'] =chunk.vacina_nome.map(lambda nome: getVaccineName(nome.upper()))
    
    #Fazendo group by pelos campos [uf, dose, mes]
    chunkToProcess = chunk.groupby(columns_group_by, as_index=False)['document_id'].count()
    
    #Chamando funcao que atualiza o dicionario com os dados do groupby feito pelo dataframe
    addItemsFromDataframeDict(brasil_data, chunkToProcess)
    
    #Fazendo group by pelos campos [uf, vacina, mes]
    chunkToProcess = chunk.groupby(['estabelecimento_uf', 'nome_vacina','vacina_mes_aplicacao'], as_index=False)['document_id'].count()
    
    #Chamando funcao que atualiza o dicionario com os dados do groupby feito pelo dataframe
    addItemsFromDataframeDictVaccine(vaccinebrasil_data, chunkToProcess)

#     if count_test > 15:
#         print("Limite para teste")
#         break;
#     count_test += 1
print("terminou")
    
```

```python
array_to_dataframe = []

for key in brasil_data:
    qtd = brasil_data[key]
    [uf, dose, mes] = key.split('_')
    #print(uf, dose, mes)
    array_to_dataframe.append({'uf': uf, 'dose': dose, 'mes' :mes, 'qtd': qtd})
#     break
```

```python
array_vaccine_to_dataframe = []

for key in vaccinebrasil_data:
    qtd = vaccinebrasil_data[key]
    [uf, vacina, mes] = key.split('_')
    #print(uf, dose, mes)
    array_vaccine_to_dataframe.append({'uf': uf, 'vacina': vacina, 'mes' :mes, 'qtd': qtd})
#     break
```

```python
brasil_df = pd.DataFrame(array_to_dataframe)
vaccine_brasil_df = pd.DataFrame(array_vaccine_to_dataframe)
```
